```python
import pandas as pd
import numpy as np
```

```python
obs_crete = pd.read_csv('crete-inaturalist-all.csv', low_memory=False)
```

```python
obs_cyprus = pd.read_csv('cyprus-inaturalist-all.csv', low_memory=False)
```

```python
paralies_df = pd.read_csv('paralies.csv', encoding='iso8859_7')

observations_near_paralies_crete = {}
observations_near_paralies_cyprus = {}

for paralia in paralies_df['Παραλία_Όνομα'].unique():
    paralia_df = paralies_df[paralies_df['Παραλία_Όνομα'] == paralia]
 
    observations_near_paralies_crete[paralia] = \
        obs_crete.loc[(obs_crete['latitude'] >= paralia_df.latitude.min()) \
        & (obs_crete['latitude'] <= paralia_df.latitude.max()) \
        & (obs_crete['longitude'] >= paralia_df.longitude.min()) \
        & (obs_crete['longitude'] <= paralia_df.longitude.max())]
    
    observations_near_paralies_cyprus[paralia] = \
        obs_cyprus.loc[(obs_cyprus['latitude'] >= paralia_df.latitude.min()) \
        & (obs_cyprus['latitude'] <= paralia_df.latitude.max()) \
        & (obs_cyprus['longitude'] >= paralia_df.longitude.min()) \
        & (obs_cyprus['longitude'] <= paralia_df.longitude.max())]
```
